# **4. Exploratory Data Analysis**
In this section I will analyse the data to answer the key questions and purpose of this data analysis.

### a) Movie Genres Analysis

Which are the top 5 most common movie genres


```python
df['Genre'].unique()
```

```python
# Split genres column into individual genre labels
genre_counts = df['Genre'].str.split(',').explode()

#  Count the occurrences of each genre
genre_counts = genre_counts.value_counts()

# Sort the genres based on their count
genre_counts_sorted = genre_counts.sort_values(ascending=False)

# Select the top 5 most common genres
top_5_genres = genre_counts_sorted.head(5)

# Display the top 5 most common genres
print(top_5_genres)
```

```python
#Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_genres.index, y=top_5_genres.values, palette='viridis')
plt.title('Top 5 Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Dramas are the most common genre followed by comedy, Action, crime and Adventure.

### d) Distribution of movies by years


```python
#Create 5 bins for years
bins = pd.cut(df['Year'], bins=5)
```

```python
#Visual representation
# Calculate the frequency of movies in each year bin
bin_counts = df['Year'].value_counts(bins=5).sort_index()

#Define colors for each bin
colors = ['skyblue', 'orange', 'green', 'red', 'purple']


# Visualize the distribution using an area plot
plt.figure(figsize=(10, 6))
bin_counts.plot(kind='area', color=colors)
plt.title('Distribution of Movies by Years (5 Bins)')
plt.xlabel('Years')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation between years binnede and revenue
# Create 5 bins for years
bins = pd.cut(df['Year'], bins=5)

# Concatenate bins with original DataFrame
df_with_bins = pd.concat([df, bins.rename('Year_Bins')], axis=1)

# Convert Year_Bins to categorical
df_with_bins['Year_Bins'] = df_with_bins['Year_Bins'].astype(str).astype('category')

# Calculate correlation between years bins and revenue
correlation = df_with_bins['Year_Bins'].cat.codes.corr(df['Revenue (Millions)'])
print("Correlation between Years Bins and Revenue:", correlation)
```

There is a negative correlation between the year and the revenue.

### e) Analyzing the revenue of the industry over the years

```python
# Bivariate Analysis of the Rvenue over the years
# Bivariate Analysis of the Revenue over the years
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
aux1 = df[['Year', 'Revenue (Millions)']].groupby('Year').sum().reset_index()
sns.barplot(x='Year', y='Revenue (Millions)', data=aux1)
plt.xticks(rotation=45)
plt.title('Total Revenue over the Years')

plt.subplot(1, 3, 2)
sns.regplot(x='Year', y='Revenue (Millions)', data=aux1)
plt.title('Regression Plot of Revenue over the Years')

plt.subplot(1, 3, 3)
sns.heatmap(aux1.corr(method='pearson'), annot=True)
plt.title('Correlation Heatmap of Revenue over the Years')

plt.tight_layout()
plt.show()
```

#### Numerical Attributes Correlation

### f) Analysis of Movie Directors

```python
#Top 10 directors generating the highest revenue

# Group the data by director and sum their revenue
director_revenue = df.groupby('Director')['Revenue (Millions)'].sum()

# Sort the results in descending order
top_directors = director_revenue.sort_values(ascending=False).head(10)

# Print the top 10 directors based on their revenue
print("Top 10 Directors based on Revenue:")
print(top_directors)
```

```python
#Visual representation of the top 10 directors
# Create a bar plot
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar', color='green', edgecolor='black')
plt.title('Top 10 Directors based on Revenue')
plt.xlabel('Director')
plt.ylabel('Total Revenue (Millions)')
plt.xticks(rotation=45)
plt.show()
```
